## task 2

```python
t = np.linspace(0, 10, num=51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')
plt.axis([0.5, 9.5, -2.5, 2.5])
# plt.xlim(0.5, 9.5)
# plt.ylim(-2.5, 2.5)
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

## task 3

```python
x = np.linspace(-3, 3, num=51)
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График y1')

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

plt.show()
```

```python
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('creditcard.csv')
class_values = df.value_counts(['Class'])
```

```python
class_values.plot(kind='bar')
plt.show()
```

```python
class_values.plot(kind='bar', logy=True)
plt.show()
```

```python
df_class_0 = df['V1'].loc[df['Class'] == 0]
df_class_1 = df['V1'].loc[df['Class'] == 1]

plt.hist(df_class_0, bins=20, density=True, label="Class 0", color="gray", alpha=0.5)
plt.hist(df_class_1, bins=20, density=True, label="Class 1", color="red", alpha=0.5)

plt.legend()
plt.xlabel('Classes')

plt.show()
```
